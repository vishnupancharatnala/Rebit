```python
import tabula
import pandas as pd
import json

# Path to the PDF file
pdf_path = "/content/hostels_permision(1).pdf"

# Output JSON file path
json_output_path = "/content/extracted_tables.json"

# Extract tables from the PDF
try:
    tables = tabula.read_pdf(pdf_path, pages="all", multiple_tables=True, lattice=True)

    # Store JSON output for all tables
    tables_json = []

    # Loop through extracted tables
    for i, table in enumerate(tables):
        print(f"Table {i+1}")
        print(table)

        # Convert each table to JSON and append to the list
        table_json = table.to_json(orient="records")
        tables_json.append(json.loads(table_json))  # Convert to Python object for better JSON handling

        # converting table to csv file
        # Save the table as a CSV
        table.to_csv(f"table_{i+1}.csv", index=False)

    # Save all tables into a single JSON file
    with open(json_output_path, "w") as json_file:
        json.dump(tables_json, json_file, indent=4)

    print(f"All tables saved successfully to {json_output_path}")

    # To download in environments like Google Colab:
    from google.colab import files
    files.download(json_output_path)

except Exception as e:
    print(f"Error extracting tables: {e}")
```
